### 2. Aggregate data from all files into one

```python
%%time
df_list = []
for i in range(1, len(files)):
    df = pd.read_csv((dpath+'{}'.format(files[i])), header = 1, skipfooter = 2, engine = 'python')
    df_list.append(df)
```

```python
len(df_list)
```

```python
%%time
df = pd.concat(df_list, ignore_index=True)
```
